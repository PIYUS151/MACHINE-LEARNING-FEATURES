# ASSIGNMENT QUESTION

**1. What is a parameter?**

->A parameter is a variable in a machine learning model that the algorithm learns from the training data (e.g., weights in linear regression).

**2.What is correlation?**

->Correlation measures the linear relationship between two variables. Values range from -1 to 1:

1: perfect positive correlation

-1: perfect negative correlation

0: no linear correlation



**3.What does negative correlation mean?**

->Negative correlation means that as one variable increases, the other decreases (e.g., hours spent watching TV and grades).

**4. Define Machine Learning. What are the main components in Machine Learning?**

->Machine Learning is a field of AI where systems learn from data to make predictions or decisions without being explicitly programmed.

Main components:

Data

Features

Model

Training

Loss function

Optimizer

Evaluation

**5. How does loss value help in determining whether the model is good or not?**

The loss value shows how far the model's predictions are from the actual values. A lower loss indicates a better model.

**6. What are continuous and categorical variables?**

Continuous: Numeric and measurable (e.g., height, temperature)

Categorical: Represent groups or categories (e.g., gender, color)

**7. How do we handle categorical variables in Machine Learning? What are the common techniques?**

->Common techniques:

  Label Encoding: Convert categories to integers

  One-Hot Encoding: Convert categories to binary columns

**8. What do you mean by training and testing a dataset?**

Training set: Used to train the model

Testing set: Used to evaluate the model's performance on unseen data

**9. What is sklearn.preprocessing?**

A module in scikit-learn for data preprocessing, including:

Scaling

Normalization

Encoding

Imputation

**10. What is a Test set?**

A test set is a subset of data used to assess how well the model generalizes to new, unseen data.

**15. What is causation? Explain the difference between correlation and causation with an example?**

->Causation means one variable directly affects another.

Example:

Correlation: Ice cream sales and drowning incidents increase in summer.

Causation: Higher temperatures cause both events, but ice cream doesn't cause drowning.

**16. What is an Optimizer? What are different types of optimizers?**

-.An optimizer adjusts model parameters to minimize the loss function.

Types:

SGD (Stochastic Gradient Descent): Simple but may be slow.

Adam: Adaptive learning rate, widely used.

RMSprop: Good for RNNs, adjusts learning rate for each parameter.
